```python
###################### 1 Задание
import pandas as pd

df = pd.read_csv("Data для тестового - Data.csv")

df["Отправлено"] = df["Отправлено"].apply(
    lambda x: int("".join(filter(str.isdigit, str(x))))
)
df["Доставлено"] = df["Доставлено"].apply(
    lambda x: int("".join(filter(str.isdigit, str(x))))
)
df["Клики"] = df["Клики"].apply(lambda x: int("".join(filter(str.isdigit, str(x)))))
df["Отписки"] = df["Отписки"].apply(lambda x: int("".join(filter(str.isdigit, str(x)))))
df["Открытия"] = df["Открытия"].apply(
    lambda x: int("".join(filter(str.isdigit, str(x))))
)
df["Пользователей на сайте"] = df["Пользователей на сайте"].apply(
    lambda x: int("".join(filter(str.isdigit, str(x))))
)

df["Дата"] = pd.to_datetime(df["Дата"])
df["Отправлено"] = df["Отправлено"].apply(lambda x: int(x))
df["Доставлено"] = df["Доставлено"].apply(lambda x: int(x))
df["Клики"] = df["Клики"].apply(lambda x: int(x))
df["Отписки"] = df["Отписки"].apply(lambda x: int(x))
df["Открытия"] = df["Открытия"].apply(lambda x: int(x))
df["Пользователей на сайте"] = df["Пользователей на сайте"].apply(lambda x: int(x))

df["Delivery rate"] = df["Доставлено"] / df["Отправлено"] * 100
df["Open rate"] = df["Открытия"] / df["Доставлено"] * 100
df["CTOR"] = df["Клики"] / df["Открытия"] * 100
df["UR"] = df["Отписки"] / df["Доставлено"] * 100

task_1 = pd.pivot_table(
    df, columns=["Дата"], index=[], values=["Delivery rate", "Open rate", "CTOR", "UR"]
)

df.to_excel('Отображение.xlsx')

task_1
###################### Конец 1 задания
```

```python
###################### 3 Задание
import pandas as pd

df = pd.read_csv("Data для тестового - Data.csv")

df["Отправлено"] = df["Отправлено"].apply(
    lambda x: int("".join(filter(str.isdigit, str(x))))
)
df["Доставлено"] = df["Доставлено"].apply(
    lambda x: int("".join(filter(str.isdigit, str(x))))
)
df["Клики"] = df["Клики"].apply(lambda x: int("".join(filter(str.isdigit, str(x)))))
df["Отписки"] = df["Отписки"].apply(lambda x: int("".join(filter(str.isdigit, str(x)))))
df["Открытия"] = df["Открытия"].apply(
    lambda x: int("".join(filter(str.isdigit, str(x))))
)
df["Пользователей на сайте"] = df["Пользователей на сайте"].apply(
    lambda x: int("".join(filter(str.isdigit, str(x))))
)

df["Дата"] = pd.to_datetime(df["Дата"])
df["Отправлено"] = df["Отправлено"].apply(lambda x: int(x))
df["Доставлено"] = df["Доставлено"].apply(lambda x: int(x))
df["Клики"] = df["Клики"].apply(lambda x: int(x))
df["Отписки"] = df["Отписки"].apply(lambda x: int(x))
df["Открытия"] = df["Открытия"].apply(lambda x: int(x))
df["Пользователей на сайте"] = df["Пользователей на сайте"].apply(lambda x: int(x))

df["Delivery rate"] = df["Доставлено"] / df["Отправлено"] * 100
df["Open rate"] = df["Открытия"] / df["Доставлено"] * 100
df["CTOR"] = df["Клики"] / df["Открытия"] * 100
df["UR"] = df["Отписки"] / df["Доставлено"] * 100

df.columns = [
    "Название рассылки",
    "Название кампании",
    "Направление",
    "Месяц",
    "Дата",
    "Год",
    "Номер недели",
    "День недели",
    "День недели.1",
    "Время",
    "Веб-версия",
    "Тема письма",
    "Сегмент",
    "Отправлено",
    "Доставлено",
    "Открытия",
    "Клики",
    "Баунсы (Все ошибки)",
    "Отписки",
    "UTM Метка",
    "Пользователей на сайте",
    "Воронка продаж. Шаг 1",
    "Воронка продаж. Шаг 2",
    "Воронка продаж. Шаг 3",
    "Delivery rate",
    "Open rate",
    "CTOR",
    "UR",
]

top_themes = (
    df.groupby("Тема письма")
    .agg(
        {
            "Открытия": "mean",
            "Клики": "mean",
            "Отправлено": "mean",
            "Пользователей на сайте": "mean",
            "Доставлено": "mean",
            "Отписки": "mean",
        }
    )
    .sort_values(
        by=[
            "Открытия",
            "Клики",
            "Отправлено",
            "Пользователей на сайте",
            "Доставлено",
            "Отписки",
        ],
        ascending=[False, False, False, False, False, True],
    )
    .reset_index()["Тема письма"]
    .values
)
print("Лучшие темы:\n", top_themes[0:5])

top_day = df.sort_values(
    by=[
        "Открытия",
        "Клики",
        "Отправлено",
        "Пользователей на сайте",
        "Доставлено",
        "Отписки",
    ],
    ascending=[False, False, False, False, False, True],
)[
    [
        "День недели.1",
        "Открытия",
        "Клики",
        "Отправлено",
        "Пользователей на сайте",
        "Доставлено",
        "Отписки",
    ]
]
top_days = (
    top_day.groupby("День недели.1")
    .agg(
        {
            "Открытия": "mean",
            "Клики": "mean",
            "Отправлено": "mean",
            "Пользователей на сайте": "mean",
            "Доставлено": "mean",
            "Отписки": "mean",
        }
    )
    .sort_values(
        by=[
            "Открытия",
            "Клики",
            "Отправлено",
            "Пользователей на сайте",
            "Доставлено",
            "Отписки",
        ],
        ascending=[False, False, False, False, False, True],
    )
    .reset_index()["День недели.1"]
    .values
)
print("Лучшие дни:\n", top_days)
###################### Конец 3 задания
```
